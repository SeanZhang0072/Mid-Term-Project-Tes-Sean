```python
# Scaling the independent variables using the Normalization or MinMaxScaler:
```

```python
from sklearn.preprocessing import MinMaxScaler

#Create an instance of MinMaxScaler:
Scaler_Type = MinMaxScaler()
# Specifiying the independent variables(features) to be scaled:
Vars_to_Scale = ['Entire home/apt', 'Shared Room', 'Private Room', 'Person Capacity', 
            'Superhost', 'Multiple Rooms', 'Business', 'Cleanliness Rating', 'Guest Satisfaction', 'Bedrooms', 
            'City Center (km)', 'Metro Distance (km)', 'Attraction Index', 'Normalised Attraction Index', 'Restraunt Index',
            'Normalised Restraunt Index']

#Scaling above features:
Scaled_Var = Scaler_Type.fit_transform(Airbnb_data[Vars_to_Scale])
Scaled_Var

# Create a DataFrame with the scaled independent variables
Scaled_data = pd.DataFrame(Scaled_Var, columns=Vars_to_Scale)

# Add the dependent variable to the DataFrame
Scaled_data['Price'] = Airbnb_data['Price']
Scaled_data['Day'] = Airbnb_data['Day']

Scaled_data



```

```python
Scaled_data.describe().T
```

Creating Multiple Regression Model with Price as a Target variable and Multiple Features as independent Variables.

```python
import statsmodels.api as sm

# Define the dependent variable (y) and independent variables (X)
y = Scaled_data['Price']
X = Scaled_data[Vars_to_Scale]

X = sm.add_constant(X) #adds a column of 1's so the model will contain an intercept

#Encode the 'Day' variable using binary encoding
encoded_data = pd.get_dummies(Scaled_data['Day'], drop_first=True)

#Combine the encoded 'Day' variable with your existing independent variables
X = pd.concat([X, encoded_data], axis=1)

#Change data type of the boolean variables in to integers to be suitable for creating the Multiple regression model.
X[['Entire home/apt', 'Shared Room', 'Private Room', 'Superhost']] = X[['Entire home/apt', 'Shared Room', 
                                                                       'Private Room', 'Superhost']].astype(int)

model = sm.OLS(y, X) #instantiate
results = model.fit() #fit the model (this is where OLS is actually being run)
print(results.summary())
```

```python
Notes:Approximately 15.5% of the variation in the Price can be explained by the independent variables included in the model.
```
